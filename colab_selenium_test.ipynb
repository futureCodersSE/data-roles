```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
#driver = webdriver.Chrome()
driver.get("https://uk.indeed.com/jobs?q=Data&l=Kent&radius=0")

#Options.headless = False
print(driver.page_source)
test1 = driver.find_element(By.ID, "jobsearch-Main") #_by_name()
LeftPane =  driver.find_element(By.CLASS_NAME,"jobsearch-LeftPane")
job_cards = LeftPane.find_elements(By.CLASS_NAME, "resultContent")
print(len(job_cards))

job_df = pd.DataFrame([])

for job in job_cards:
    #job = job_cards
    try:
        job_title = job.find_element(By.TAG_NAME,"span").text    #class= "jobTitle jobTitle-newJob css-bdjp2 meu4oa1s0 tabindex="-1"
        
    except:
        job_title = ""
    
    try:
        job_company_name = job.find_element(By.CLASS_NAME, "companyName").text #class = "attribute_snippet
        
    except:
        job_company_name = ""
        
    try:    
        job_location = job.find_element(By.CLASS_NAME, "companyLocation").text
        
    except:
        job_location = ""
        
    try:   
        job_pay = job.find_element(By.CLASS_NAME,"attribute_snippet").text
        
    except:
        job_pay = ""
        
    try:   
        job_contract = job.find_elements(By.CLASS_NAME, "metadata")[1].text

    except:
        job_contract = ""
        
    try:
        job_url = job.find_element(By.TAG_NAME, "a").get_attribute("href") #(By.CLASS_NAME, "jcs-JobTitle").get_attribute("href")

    except:
        print(job_url = "")
    print("job_title: ",job_title)    
    print("job_company: ",job_company_name)
    print("job_location: ",job_location)
    print("job_pay: ",job_pay)   
    print("job_contract: ",job_contract)  
    print("job_url", job_url)
    print("\n\n")
    
    new_row = {"job_title":job_title, "company":job_company_name, "salary": job_pay, "contract":job_contract,
               "location":job_location, "job_url":job_url }
    
    job_df = job_df.append(new_row,ignore_index=True)


#print(driver.page_source)

driver.quit()
print(job_df)
```
